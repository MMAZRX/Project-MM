```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split    
from sklearn.svm import SVC
from sklearn.metrics import classification_report  
from sklearn.metrics import confusion_matrix, accuracy_score

url = 'D:\\__Max__\\Documents\\Machine Learning Course LANIA\\Recursos complementarios-20210804\\Datasets\\Failure prediction.csv'
failure = pd.read_csv(url)
print(failure.shape)
print(failure.head())
```

```python
propiedades = failure.iloc[:,1:]
objetivo = failure['GOAL-Broken']
print(propiedades.head())
print(objetivo.head())
```

```python
propiedades_train, propiedades_test, objetivo_train, objetivo_test = train_test_split(propiedades, objetivo, test_size=0.2, random_state=0)
print('Número de instancias de entrenamiento: '+str(len(propiedades_train)))
print('Número de instancias de pureba: '+str(len(propiedades_test)))
print('Número de instancias totales: '+str(len(failure.index)))
```

## Algoritmo: Máquina de Soporte Vectorial Clasificación para la predicción de fallas.

```python
modelo = SVC(kernel = 'rbf')
modelo.fit(propiedades_train, objetivo_train)
```

```python
prediccion_objetivo = modelo.predict(propiedades_test)
```

```python
print(classification_report(objetivo_test, prediccion_objetivo))
```

```python
MC = confusion_matrix(objetivo_test, prediccion_objetivo)
print(MC)
print(accuracy_score(objetivo_test, prediccion_objetivo))
```

```python
sns.heatmap(MC, annot=True, cbar=True, cmap="coolwarm")
plt.show()
```

```python
precision = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
    knn.fit(propiedades_train, objetivo_train)
    pred_i = knn.predict(propiedades_test)
    precision.append(accuracy_score(objetivo_test, pred_i))
```

## Reducción de Dimensionalidad. PCA

A continuación se aplicará el preocedimiento de Análisis de Componente Principal (PCA) para reducir la dimensionalidad del Dataset Failure Prediction (Predicción de Fallas). Para comparar resultados como parte de la Práctica no.4

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

urlPCA = 'D:\\__Max__\\Documents\\Machine Learning Course LANIA\\1. Fundamentos Machine Leargning\\Datasets\\Failure prediction.csv'
failurePCA = pd.read_csv(urlPCA)
print(failurePCA.shape)
print(failurePCA.head())
```

```python
features = failurePCA.iloc[:,1:]
target = failurePCA['GOAL-Broken']
features.hist(bins=50, figsize=(15,15))
plt.show()
```

```python
target.hist(bins=3, figsize=(2,2))
plt.show()
```

Después de cargar el Dataset, se procede a realizar una Estandarización de los Datos, ya que esto complementa y mejora la Reducción de Dimensionalidad por PCA.

```python
scaler = StandardScaler()
featuresSTD = scaler.fit_transform(features)
featuresSTD
```

Una vez realizada la Estandarización, procedemos a aplicar el método PCA para realizar la reducción de dimensionalidad del DataSet.

```python
pca = PCA()
pca.fit_transform(featuresSTD)
pca_variance = pca.explained_variance_

pca_variance
```

```python
plt.figure(figsize=(8,6))
plt.bar(range(8), pca_variance, alpha=0.5, align='center', label="Varianza Individual")
plt.legend()
plt.ylabel('Varianza')
plt.xlabel('Componentes Principales')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
targetCodif = encoder.fit_transform(target)
targetCodif
```

```python
pca2 = PCA(n_components=7)
pca2.fit(featuresSTD)
x = pca2.transform(featuresSTD)

plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=targetCodif)
plt.show()
```

```python
pca2 = PCA(n_components=2)
pca2.fit(featuresSTD)
x = pca2.transform(featuresSTD)

plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=targetCodif)
plt.show()
```

```python
def get_pca_components(pca, var):
    cumm_var = pca.explained_variance_ratio_
    total_var = 0
    N = 0
    for i in cumm_var:
        N += 1
        total_var += i
        if total_var >= var:
            break
    return N
```

```python
pca3 = PCA().fit(featuresSTD)
n = get_pca_components(pca3, 0.95)
n
```
